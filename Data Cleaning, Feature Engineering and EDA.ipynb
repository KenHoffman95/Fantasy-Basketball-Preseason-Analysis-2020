# Fantasy Basketball 2020 Draft Guide

## Data Cleaning

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_excel('2019-2020 NBA Stats.xlsx')
```

```python
df.head(25)
```

```python
df.shape
```

```python
#drop unnecessary columns
df = df.drop('Rk', 1)
df = df.drop('Tm', 1)
df = df.drop('GS', 1)
df = df.drop('3PA', 1)
df = df.drop('3P%', 1)
df = df.drop('2P', 1)
df = df.drop('2PA', 1)
df = df.drop('2P%', 1)
df = df.drop('eFG%', 1)
df = df.drop('ORB', 1)
df = df.drop('DRB', 1)
df = df.drop('PF', 1)
```

```python
df.head(250)
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
#remove anyone with a NA FT% or FG% so their lack of playing time and stat accumulation won't effect the overall data
df = df.dropna(how='any',axis=0) 
```

```python
#remove anyone with a 1.00 FG% so their lack of playing time and stat accumaltion won't effect the overall data
df = df[df['FG%'] != 1.00]
```

```python
df.shape
```

```python
#Rename 'TRB' column to 'REB'
df = df.rename(columns={'TRB': 'REB'})
```

```python
df.info()
```

```python
#Update players with multiple positions to only include main position
df['Pos'] = df['Pos'].map({'C': 'C', 'C-PF': 'C', 'PF': 'PF', 'PF-C': 'PF', 'PF-SF': 'PF', 'PG': 'PG', 
                           'PG-SG': 'PG', 'SF': 'SF', 'SF-C': 'SF', 'SF-PF': 'SF', 'SF-SG': 'SF',
                           'SG': 'SG', 'SG-PG': 'SG','SG-SF': 'SG'})
```

# Feature Engineering

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
#create column Double Doubles per Game
df['DD_per'] = df['DD']/df['G']
```

```python
#create column Triple Doubles per Game
df['TD_per'] = df['TD']/df['G']
```

```python
#create column Turnovers per minute to more accurately account for this stat
df['TOVPM'] = df['TOV']/df['MP']
```

### Building Fantasy Value Metric

To this point, I've organized each player into a tier for the following 9 statistics: FG%, 3PM, FT%, REB, AST, STL, BLK, TOV, and PTS. I've created two different ways to organize players into tiers by statistic (Overall and by Position). I will now sum each player's tiers up to build an overall fantasy value metric. While performing EDA, I realized that the statistics of the lower-end players are affecting my ability to build meaningful insights, as most of the top players and 2nd tier players are receiving 3s in every category. After building the fantasy value metric, I will remove the bottom 200 players (the ones who, for the most part, won't be drafted in fantasy basketball), and then repeat the process of assigning tiers and building an overall metric.

```python
#create Fantasy Value Metric (League-Wide)
df['FV_Overall'] = df['FG_Tier'] + df['3P_Tier'] + df['FT_Tier'] + df['REB_Tier'] + df['AST_Tier'] + df['STL_Tier'] + df['BLK_Tier'] + df['TOV_Tier'] + df['PTS_Tier']
```
