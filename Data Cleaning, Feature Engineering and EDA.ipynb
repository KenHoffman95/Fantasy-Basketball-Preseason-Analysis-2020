# Fantasy Basketball 2020 Draft Guide

## Data Cleaning

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_excel('2019-2020 NBA Stats.xlsx')
```

```python
df.head(25)
```

```python
df.shape
```

```python
#drop unnecessary columns
df = df.drop('Rk', 1)
df = df.drop('Tm', 1)
df = df.drop('GS', 1)
df = df.drop('3PA', 1)
df = df.drop('3P%', 1)
df = df.drop('2P', 1)
df = df.drop('2PA', 1)
df = df.drop('2P%', 1)
df = df.drop('eFG%', 1)
df = df.drop('ORB', 1)
df = df.drop('DRB', 1)
df = df.drop('PF', 1)
```

```python
df.head(250)
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
#remove anyone with a NA FT% or FG% so their lack of playing time and stat accumulation won't effect the overall data
df = df.dropna(how='any',axis=0) 
```

```python
#remove anyone with a 1.00 FG% so their lack of playing time and stat accumaltion won't effect the overall data
df = df[df['FG%'] != 1.00]
```

```python
df.shape
```

```python
df.info()
```

```python
#Update players with multiple positions to only include main position
df['Pos'] = df['Pos'].map({'C': 'C', 'C-PF': 'C', 'PF': 'PF', 'PF-C': 'PF', 'PF-SF': 'PF', 'PG': 'PG', 
                           'PG-SG': 'PG', 'SF': 'SF', 'SF-C': 'SF', 'SF-PF': 'SF', 'SF-SG': 'SF',
                           'SG': 'SG', 'SG-PG': 'SG','SG-SF': 'SG'})
```

# Feature Engineering

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
#create column Double Doubles per Game
df['DD_per'] = df['DD']/df['G']
```

```python
#create column Triple Doubles per Game
df['TD_per'] = df['TD']/df['G']
```
