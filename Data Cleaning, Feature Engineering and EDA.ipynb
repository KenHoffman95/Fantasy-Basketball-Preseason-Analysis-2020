```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_excel('2019-2020 NBA Stats.xlsx')
df1 = pd.read_excel('2019-2020 Basketbal DD and TD.xlsx')
```

```python
df.head(25)
```

```python
df.shape
```

```python
#Merge df_nba and df_ncaa
df = pd.merge(df, df1, on='Player', how='inner')
#df = pd.merge(df,df_combine,on='Player', how='outer')
```

```python
df.head(5)
```

```python
#drop unnecessary columns
df = df.drop('Rk_x', 1)
df = df.drop('3PA', 1)
df = df.drop('3P%', 1)
df = df.drop('2P', 1)
df = df.drop('2PA', 1)
df = df.drop('2P%', 1)
df = df.drop('eFG%', 1)
df = df.drop('ORB', 1)
df = df.drop('DRB', 1)
df = df.drop('PF', 1)
df = df.drop('Rk_y', 1)
```
